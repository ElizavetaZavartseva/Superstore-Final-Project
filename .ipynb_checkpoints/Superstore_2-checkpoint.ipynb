```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 500)

pd.set_option('display.max_colwidth', 1)
#to show all values in row
```

```python
df_superstore = pd.read_csv('superstore.csv')
df_superstore
```

# Data Cleaning

```python
#Checking duplicated values

df_superstore.duplicated().sum()
```

```python
#Checking null values

df_superstore.isna().sum()
```

```python
#Renaiming columns

df_superstore.columns = df_superstore.columns.str.lower()\
                                                    .str.replace(' ', '_')\
                                                    .str.replace('.', '_')\
                                                    .str.replace('order_id', 'order_number')\
                                                    .str.replace('product_id', 'product_number')
```

```python
#Removing some columns

df_superstore.drop(['记录数', 'year', 'market2', 'weeknum', 'row_id'], axis=1, inplace=True)
df_superstore.head(1)
```

### < Convert object columns to datetime64

```python
df_superstore.dtypes
```

```python
#Cleaning column "order_date"

df_superstore["order_date"] = df_superstore["order_date"].str.split().str[0]
df_superstore["order_date"] = pd.to_datetime(df_superstore["order_date"], format="%Y/%m/%d")

#To pull year from data: df_superstore['order_date'].dt.year
```

```python
#Cleaning column "ship_date"

df_superstore["ship_date"] = df_superstore["ship_date"].str.split().str[0]
df_superstore["ship_date"] = pd.to_datetime(df_superstore["ship_date"], format="%Y/%m/%d")

df_superstore.head(1)
```

```python
df_superstore.dtypes
```

```python
df_superstore.shape
```

# < Creating new columns for creating DataBase

### 'oreder_id' 

```python
df_superstore['cum_count'] = df_superstore.groupby('order_number').cumcount() + 1
```

```python
df_superstore['order_id'] = df_superstore['order_number'] + '_' + df_superstore['cum_count'].astype(str)
df_superstore.head()
```

### 'market_id' 

```python
df_market_ids = df_superstore[['market', 'country']]
df_market_ids = df_market_ids.drop_duplicates()
df_market_ids['country_number'] = df_market_ids.groupby(['market']).cumcount() + 1
df_market_ids
```

```python
df_superstore = pd.merge(df_superstore, df_market_ids, on=['market', 'country'], how='left')
df_superstore['market_id'] = df_superstore['market'] + '-' + df_superstore['country_number'].astype(str)
df_superstore.head(1)
```

```python
df_superstore.drop(['cum_count', 'country_number'], axis=1, inplace=True)
df_superstore.head(1)
```

```python
df_superstore.nunique()
```

```python
def SQL_INSERT_STATEMENT_FROM_DATAFRAME(SOURCE, TARGET):
    sql_texts = []
    for index, row in SOURCE.iterrows():       
        sql_texts.append('INSERT INTO '+TARGET+' ('+ str(', '.join(SOURCE.columns))+ ') VALUES '+ str(tuple(row.values)))        
    return sql_texts

def SQL_CREATE_STATEMENT_FROM_DATAFRAME(SOURCE, TARGET):
    # SQL_CREATE_STATEMENT_FROM_DATAFRAME(SOURCE, TARGET)
    # SOURCE: source dataframe
    # TARGET: target table to be created in database

    sql_text = pd.io.sql.get_schema(SOURCE.reset_index(), TARGET)   
    return sql_text

user = 'root'
password = 'Qwerty123'
host = 'localhost'
database = 'superstore_db'

db = mysql.connector.connect(
    user=user,
    password=password,
    host=host,
    database=database
)

# from sqlalchemy import create_engine
# engine = create_engine(f"mysql://root:Qwerty123@localhost/youtube_db")

insert_statements_list = SQL_INSERT_STATEMENT_FROM_DATAFRAME(df_customers, "customers")
text_query = SQL_CREATE_STATEMENT_FROM_DATAFRAME(df_customers, "customers")

print(text_query)
```
